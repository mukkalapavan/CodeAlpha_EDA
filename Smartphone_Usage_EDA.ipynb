```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Smartphone_Usage.csv")
df.shape

```

```python
#To print first 5 rows
df.head()
```

```python
#Information and Types
df.info()
df.dtypes
```

```python
#Statitical Values
df.describe()
```

```python
#To find number of null values
df.isnull()
```

```python
#Null values count by Column
df.isnull().sum()
```

```python
#Handling Missing Values
df.fillna(0,inplace=True)
```

```python
df
```

```python
#For Duplicate Values
df.duplicated().sum()
```

```python
#correlation matrix
plt.figure()
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
plt.title("correlation Matrix")
plt.show()
```

```python
#Age Dustribution
sns.histplot(df["Age"],kde=True)
plt.title("Age Distribution")
```

```python
plt.show()
```

```python
#Productivity vs Phone Usage
sns.scatterplot(x="Daily_Phone_Hours", y="Work_Productivity_Score", data=df)
plt.title("Phone Usage vs Productivity")
plt.show()
```

```python
#Gender-wise Phone Usage
sns.boxplot(x="Gender", y="Daily_Phone_Hours", data=df)
plt.title("Gender vs Daily Phone Usage")
plt.show()
```

```python
#Device Type Count
df["Device_Type"].value_counts().plot(kind="bar")
plt.title("Device Type Distribution")
plt.xticks(rotation=45)
plt.show()
```
